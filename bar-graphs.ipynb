```python
import matplotlib.pyplot as plt
import pandas as pd
import csv

import load as ld
```

```python
def draw_bar_graph_1(dataset: pd.DataFrame, cols: list) -> None:
    x = dataset[dataset[cols[1]] > 8][cols[0]][::50]
    y = dataset[dataset[cols[1]] > 8][cols[1]][::50]
    plt.bar(x, y)
    plt.legend()
    plt.show()
```

```python
anime_dataset = ld.getData(ld.ANIME_DATASET_PATH)
family_dataset = ld.getData(ld.FAMILIES_DATASET_PATH)
air_dataset = ld.getData(ld.AIR_DATASET_PATH)
bmi_dataset = ld.getData(ld.BMI_DATASET_PATH)
```

```python
draw_bar_graph_1(anime_dataset, ["ranked", "score"])
```
